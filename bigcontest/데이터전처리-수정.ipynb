```python
# df2에 무형데이터 제외한 데이터 삽입
idx_fl = df[df['상품군'] == "무형"].index
df2 = df.drop(idx_fl)
df2
```

```python
# 판매단가 및 취급액 데이터 앞뒤공백제거
df2['판매단가']  = df2['판매단가'].str.strip()
df2['취급액']  = df2['취급액'].str.strip()
df2
```

```python
# 판매단가 및 취급액 데이터 앞뒤공백, ","제거
df2['판매단가']  = df2['판매단가'].str.replace(",","")
df2['취급액']  = df2['취급액'].str.replace(",","")
df2
```

```python
df2.dtypes
```

```python
#판매단가 및 취급액 데이터 int로 변환
df2[['판매단가','취급액']] = df2[['판매단가','취급액']].apply(pd.to_numeric)
df2
```

```python
df2.dtypes
```

```python
from matplotlib import pyplot as plt
```

```python
#방송일시를 datetime object 로 변경
df2['방송일시'] = df2['방송일시'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d %H:%M'))
df2
```

```python
df2.columns = ['datetime', 'exp', 'mcode', 'pcode', 'name', 'type', 'price', 'amount','sale']
df2
```

# 1. 중복된 시간을 합친다. 
# 2. 합치면서 마더코드, 상품군, 가격, 노출시간은 유지하되,
# 취급액, 판매수량은 더한다.

```python
# 방송일시(기준열), 노출시간, 마더코드, 상품군, 가격 (기준에 따라 합쳐질 때 변하지 않아야 할 열) 추출 
df2_sample = df2[['datetime','exp','mcode','type','price']]
df2_sample
```

```python
ddf=df2.groupby('datetime').sum() #방송일시별로 모든 연속형 변수 합함.
ddf2=ddf[['amount','sale']] # 방송일시별 취급액, 판매수량의 합
ddf3=ddf2.reset_index(level='datetime') # 방송일시(index)를 column으로 변환
ddf3
```

```python
# datetime별로 중복되는 행의 취급액과 판매수량이 합쳐졌지만, 중복되는 결과 나타남.
ddf4 = pd.merge(df2_sample,ddf3)
ddf4
```

```python
# datetime을 기준으로 중복되는 행 제거.
ddf5=ddf4.drop_duplicates(['datetime'])
ddf5
```

---------------------------------------------------------------------------------------------
# 여기까지 방송일시가 중복되는 행 병합

```python
#시간 제외 데이터 추출하여 'date'변수로 할당
ddf5['date'] = ddf5['datetime'].dt.strftime('%Y%m%d')
ddf5
```

```python
# 마더코드와 날짜, 가격에 따른 노출시간, 취급액, 판매수량의 합
mdf=ddf5.groupby(['date','mcode','type','price']).sum()
mdf2=mdf[['exp','amount','sale']]
mdf2
```

```python
# index를 column으로 바꾸기
mdf3=mdf2.reset_index(level=['date', 'mcode','type','price'])
mdf3
```

```python
# 위의 데이터는 방송일자만 고려함. 방송시간또한 고려해야하므로 datetime변수를 위 데이터와 merge해야함.
# 마더코드, date, 판매단가를 고려하여 방송일시 열을 합쳐야 함 - 세가지중 하나라도 빠지면 취급액과 판매수량이 달라짐.
ddf6=ddf5[['datetime','mcode','date','price']]
ddf6
```

```python
# 중복열을 기준으로 merge
ddf7 = pd.merge(ddf6,mdf3)
ddf7
```

```python
# 마더코드, date, 판매단가가 중복되는 행만 제거하고 첫번째 행만 남게 함.
ddf8=ddf7.drop_duplicates(['mcode','date','price'])
ddf8
```

```python
#이제 쓸모없어진 date열 제거
del ddf8['date']
```

```python
# 원래대로 열 재배치
ddf8 = ddf8[['datetime','exp','mcode','type','price', 'amount', 'sale']]
ddf8
```
