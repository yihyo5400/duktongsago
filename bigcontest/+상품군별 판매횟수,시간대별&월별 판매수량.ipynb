```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
df=pd.read_csv("2019.csv")
df.head()
```

```python
import time
import datetime
```

```python
df["노출(분)"]=df["노출(분)"].fillna(method='ffill')
df
```

```python
df.rename(columns={' 취급액 ':'취급액'}, inplace = True)
```

```python
dfformless = df[df['상품군'] == "무형"]
dfformless
```

```python
idx_fl = df[df['상품군'] == "무형"].index
df2 = df.drop(idx_fl)
df2
```

```python
# 판매단가 및 취급액 데이터 앞뒤공백제거
df2['판매단가']  = df2['판매단가'].str.strip()
df2
```

```python
df.rename(columns={' 취급액 ':'취급액'}, inplace = True)
# 판매단가 및 취급액 데이터 앞뒤공백제거
df2['판매단가']  = df2['판매단가'].str.strip()
df2['취급액']  = df2['취급액'].str.strip()
df2
```

```python
df2['판매단가']  = df2['판매단가'].str.replace(",","")
df2['취급액']  = df2['취급액'].str.replace(",","")
df2
```

```python
df2[['판매단가','취급액']] = df2[['판매단가','취급액']].apply(pd.to_numeric)
df2
```

```python
from matplotlib import pyplot as plt
```

```python
#방송일시를 datetime object 로 변경
df2['방송일시'] = df2['방송일시'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d %H:%M'))
df2
```

```python
df2.set_index(df2['방송일시'], inplace=True)
df2
```

```python
df2['판매수량'] = df2['취급액']/df2['판매단가']
df2
```

```python
# 시간별 판매수량의 합
df2.판매수량.resample('H').sum()
```

```python
#시간별 판매수량
df2_1=df2.판매수량.resample('H').sum()
DataFrame(df2_1)
```

```python
#월별 판매수량
df2_2=df2.판매수량.resample('1M').sum()
DataFrame(df2_2)

```

```python
df2_2[0]
```

```python
#월별 판매수량 그래프
plt.plot(df2_2)
```

```python
#상품군별 판매수량
q=df2['판매수량'].groupby(df2['상품군']).sum()

DataFrame(q)
```
