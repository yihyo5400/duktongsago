```python
%matplotlib inline

import pandas as pd
import numpy as np
from scipy import stats
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns

import time
import datetime
```

```python
df=pd.read_csv("2019.csv",encoding='CP949')
df.head()
```

```python
df.columns = ['datetime', 'exp', 'mcode', 'pcode', 'name', 'type', 'price', 'amount']
df
```

```python
df.dtypes
```

```python
# 노출시간 NaN값 채우기

df["exp"]=df["exp"].fillna(method='ffill')
df
```

```python
# 무형데이터 삭제

idx_fl = df[df['type'] == "무형"].index
df = df.drop(idx_fl)
df
```

```python
# 결측값 있는 행 확인

df.isnull().sum()
```

```python
# 결측값 제거

df=df.dropna(axis=0)
df
```

```python
df.isnull().sum()
```

```python
# price 와 amount데이터를 str이 아닌 숫자형으로 바꾸기 위한 밑작업 - 통화기호 ","제거

df['price']  = df['price'].str.replace(",","")
df['amount']  = df['amount'].str.replace(",","")
df
```

```python
df[['price','amount']] = df[['price','amount']].astype('int')
df
```

```python
#방송일시를 datetime object 로 변경

df['datetime'] = df['datetime'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d %H:%M'))
df
```

```python
df['sale'] = df['amount']/df['price']
df
```

# 중복되는 시간 합치기-상품명 고려하지 않음 

```python
# datetime중복되는 행 합치기 
# 방송일시(기준열), 노출시간, 마더코드, 상품군 (기준에 따라 합쳐질 때 변하지 않아야 할 열) 추출 

df_sample = df[['datetime','exp','mcode','type']]
df_sample
```

```python
# 판매단가는 각상품들의 평균으로 계산하기

df.groupby('datetime').price.mean()
```

```python
# datetime별로 amount와 sale 변수 합함
dfc1=df.groupby('datetime')['amount','sale'].sum().reset_index(level='datetime')
# datetime별로 price의 평균 구함.
dfc2=df.groupby('datetime').price.mean().reset_index(level='datetime')
# 두 데이터 병합
dfc = pd.merge(dfc2,dfc1)
dfc
```

```python
df_sp2 = pd.merge(df_sample,dfc)
df_sp2

# datetime별로 중복되는 행의 취급액과 판매수량이 합쳐졌지만, 중복되는 결과 나타남.
```

```python
# datetime을 기준으로 중복되는 행 제거.
df_sp3=df_sp2.drop_duplicates(['datetime'])
df_sp3
```

```python
# 분 제외 데이터 추출하여 'date'변수로 할당 - 연속되는 시간의 동일한 상품의 취급액과 판매수량을 합치기 위함. 
# ex) 2019-01-01 06:00:00 / 60 / 100346 / 의류 / 39900 / 6470000+10217000+16009000 / 162.155388+256.065163+401.228070

df_sp3['date'] = df_sp3['datetime'].dt.strftime('%Y%m%d%H')
df_sp3
```

```python
# mcode와 date, price에 따른 exp, amount, sale의 합

df_sp4=df_sp3.groupby(['date','mcode','type','price'])['exp','amount','sale'].sum()
df_sp4
```

```python
# index를 column으로 바꾸기
df_sp4=df_sp4.reset_index(level=['date', 'mcode','type','price'])
df_sp4
```

```python
# 방송시작시간을 넣어야하므로 datetime변수를 위 데이터와 merge필요.
# mcode(price), date를 고려하여 방송일시 열을 합쳐야 함

df_sp5=df_sp3[['datetime','mcode','date','price']]
df_sp5
```

```python
# 중복열을 기준으로 merge
df_sp6 = pd.merge(df_sp5,df_sp4)
df_sp6
```

```python
# 마더코드, date, 판매단가가 중복되는 행만 제거하고 첫번째 행만 남게 함.
df_sp6=df_sp6.drop_duplicates(['mcode','date','price'])
df_sp6
```

```python
# 16:20~17:20 이렇게 20분을 기준으로 한시간 단위로 편성된 경우 존재. - 시간을 추출한 부분의 오류
df_sp6.tail(50)
```

```python
# mcode가 중복된다면, 중복 행의 exp, amount, sale을 합해서 중복되는 첫번째 행에 할당.

for i in range(1,len(df_sp6)):
    if df_sp6['mcode'].iloc[i] == df_sp6['mcode'].iloc[i-1] :
        df_sp6['exp'].iloc[i-1] = df_sp6['exp'].iloc[i-1] + df_sp6['exp'].iloc[i]
        df_sp6['amount'].iloc[i-1] = df_sp6['amount'].iloc[i-1] + df_sp6['amount'].iloc[i]
        df_sp6['sale'].iloc[i-1] = df_sp6['sale'].iloc[i-1] + df_sp6['sale'].iloc[i]
df_sp6
```

```python
# 중복된 첫 행만 남기고 제거

df_sp7=df_sp6.drop(df_sp6[df_sp6['mcode'] == df_sp6['mcode'].shift(1)].index)
df_sp7
```

```python
#이제 쓸모없어진 date열 제거
del df_sp7['date']
```

```python
# 원래대로 열 재배치
shopping = df_sp7[['datetime', 'exp', 'mcode', 'type', 'price', 'amount', 'sale']]
shopping
```

```python
# float to int
shopping=shopping.astype({'exp': int, 'price': int})
shopping
```
