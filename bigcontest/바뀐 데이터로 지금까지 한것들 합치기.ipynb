```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
df=pd.read_csv("2019.csv")
df.head()
```

```python
import time
import datetime
```

```python
df["노출(분)"]=df["노출(분)"].fillna(method='ffill')
df
```

```python
df.rename(columns={' 취급액 ':'취급액'}, inplace = True)
```

```python
dfformless = df[df['상품군'] == "무형"]
dfformless
```

```python
idx_fl = df[df['상품군'] == "무형"].index
df2 = df.drop(idx_fl)
df2
```

```python
# 판매단가 및 취급액 데이터 앞뒤공백제거
df2['판매단가']  = df2['판매단가'].str.strip()
df2
```

```python
df.rename(columns={' 취급액 ':'취급액'}, inplace = True)
# 판매단가 및 취급액 데이터 앞뒤공백제거
df2['판매단가']  = df2['판매단가'].str.strip()
df2['취급액']  = df2['취급액'].str.strip()
df2
```

```python
df2['판매단가']  = df2['판매단가'].str.replace(",","")
df2['취급액']  = df2['취급액'].str.replace(",","")
df2
```

```python
df2[['판매단가','취급액']] = df2[['판매단가','취급액']].apply(pd.to_numeric)
df2
```

```python
from matplotlib import pyplot as plt
```

```python
#방송일시를 datetime object 로 변경
df2['방송일시'] = df2['방송일시'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d %H:%M'))
df2
```

```python
# 방송일시(기준열), 노출시간, 마더코드, 상품군, 가격 (기준에 따라 합쳐질 때 변하지 않아야 할 열) 추출 
df2_sample = df2[['방송일시','노출(분)','마더코드','상품군','판매단가']]
df2_sample
```

```python
df2['판매수량'] = df2['취급액']/df2['판매단가']
df2
```

```python
ddf=df2.groupby('방송일시').sum() #방송일시별로 모든 연속형 변수 합함.
ddf2=ddf[['취급액','판매수량']] # 방송일시별 취급액, 판매수량의 합
ddf3=ddf2.reset_index(level='방송일시') # 방송일시(index)를 column으로 변환
ddf3
```

```python
# datetime별로 중복되는 행의 취급액과 판매수량이 합쳐졌지만, 중복되는 결과 나타남.
ddf4 = pd.merge(df2_sample,ddf3)
ddf4
```

```python
# datetime을 기준으로 중복되는 행 제거.
ddf5=ddf4.drop_duplicates(['방송일시'])
ddf5
```

```python
#시간 제외 데이터 추출하여 'date'변수로 할당
ddf5['날짜'] = ddf5['방송일시'].dt.strftime('%Y%m%d')
ddf5
```

```python
# 마더코드와 날짜, 가격에 따른 노출시간, 취급액, 판매수량의 합
mdf=ddf5.groupby(['날짜','마더코드','상품군','판매단가']).sum()
mdf2=mdf[['노출(분)','취급액','판매수량']]
mdf2
```

```python
# index를 column으로 바꾸기
mdf3=mdf2.reset_index(level=['날짜', '마더코드','상품군','판매단가'])
mdf3
```

```python
# 위의 데이터는 방송일자만 고려함. 방송시간또한 고려해야하므로 datetime변수를 위 데이터와 merge해야함.
# 마더코드, date, 판매단가를 고려하여 방송일시 열을 합쳐야 함 - 세가지중 하나라도 빠지면 취급액과 판매수량이 달라짐.
ddf6=ddf5[['방송일시','마더코드','날짜','판매단가']]
ddf6
```

```python
# 중복열을 기준으로 merge
ddf7 = pd.merge(ddf6,mdf3)
ddf7
```

```python
# 마더코드, date, 판매단가가 중복되는 행만 제거하고 첫번째 행만 남게 함.
ddf8=ddf7.drop_duplicates(['마더코드','날짜','판매단가'])
ddf8
```
